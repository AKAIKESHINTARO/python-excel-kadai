```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

df_2022 = pd.read_excel(
    "2022_年間売上表.xlsx",
    sheet_name="Sheet1"
)

df_2023 = pd.read_excel(
    "2023_年間売上表.xlsx",
    sheet_name="Sheet1"
)

df_all = pd.concat(
    [df_2022, df_2023],
    ignore_index=True
)

result = (
    df_all
    .groupby(["商品", "売上年"])["金額（千円）"]
    .sum()
    .reset_index()
)

output_file = "売上集計表.xlsx"
result.to_excel(output_file, index=False)

wb = load_workbook(output_file)
ws = wb.active

header_fill = PatternFill(
    fill_type="solid",
    fgColor="F2F2F2"
)

for cell in ws[1]:
    cell.fill = header_fill

wb.save(output_file)
```
